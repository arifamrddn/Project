```
# Load libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.naive_bayes import GaussianNB
```

```
col_names = ['Car Brand',
             #'Transmission', 
             'Car Price', 
             'Loan duration', 
             'Sex', 
             'Age', 
             'Type Of Employment', 
             'Car Type']

# load dataset
dataset = pd.read_csv("/content/Customer Feedback (Responses) - Form responses 1.csv")
dataset
```

```
#data cleaning
#convert data string into integer
# Transmission attributes not converted because it is not in use 
d = {'Perodua': 1, 'Honda': 2, 'Toyota': 3, 'Proton': 4}
dataset['Car Brand'] = dataset['Car Brand'].map(d)

d = {'Less than RM 60 000': 1, 'More than RM 60 000': 0}
dataset['Car Price'] = dataset['Car Price'].map(d)

d = {'More than 5 years': 1, '5 Years': 0}
dataset['Loan duration '] = dataset['Loan duration '].map(d)

d = {'Male': 1, 'Female': 0}
dataset['Sex'] = dataset['Sex'].map(d)

d = {'20 - 30': 1, 'More than 30': 0}
dataset['Age'] = dataset['Age'].map(d)

d = {'Employed': 1, 'Unemployed': 0}
dataset['Type Of Employment'] = dataset['Type Of Employment'].map(d)

d = {'Perodua Axia': 1, 
     'Perodua Myvi': 2, 
     'Honda HRV': 3, 
     'Toyota Camry': 4, 
     'Honda Jazz': 5, 
     'Proton Persona': 6, 
     'Proton Iriz': 7, 
     'Perodua Bezza': 8}
dataset['Car Type'] = dataset['Car Type'].map(d)
dataset
```

```
#split dataset in features and target variable
feature_cols = [ 'Car Brand',
                #'Transmission ', 
                'Car Price', 
                'Loan duration ', 
                'Sex', 
                'Age',
                'Type Of Employment']

X = dataset[feature_cols] # Features
y = dataset['Car Type'] # Target variable
```

```
#Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=100) # 80% training and 20% test
```

```
# Using Gausian Naive Bayes classifier
model = GaussianNB()
model.fit(X_train,y_train)
# Calculate the accuracy of using Naive Bayes classifier 
print("Accuracy using Naive Bayes:",metrics.accuracy_score(y_test, model.predict(X_test))*100, "%")
```

```
# predict output given the attributes value from features
predicted= model.predict([[4,0,0,1,1,0]]) # car brand , car price, loan duration, sex, age, type of employment
print("Predicted Value is:", predicted)
print("Therefore, the suggested car would be")
if predicted ==1:
  print("Perodua Axia")
if predicted ==2:
  print("Perodua Myvi")
if predicted ==3:
  print("Honda HRV")
if predicted ==4:
  print("Toyota Camry")
if predicted ==5:
  print("Honda Jazz")
if predicted ==6:
  print("Proton Persona")
if predicted ==7:
  print("Proton Iriz")
if predicted ==8:
  print("Perodua Bezza")
```
